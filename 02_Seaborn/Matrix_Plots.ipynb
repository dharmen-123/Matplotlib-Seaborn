## ************** Matrix Plots **************

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
print("Import successfully")
```

```python
flights=sns.load_dataset('flights')
flights
```

```python
tips=sns.load_dataset('tips')
tips
```

### Heat Map

```python
tipcorr=tips[['total_bill','tip','size']]
tipcorr
```

```python
df1=tipcorr.corr()
df1
```

```python
df2=flights[['year','passengers']].corr()
df2
```

```python
sns.heatmap(df1,annot=True)
```

```python
sns.heatmap(df2,annot=True,linecolor='red',linewidths=2)
```

### Cluster map

```python
sns.clustermap(df1,figsize=(7,4))
```

```python
sns.clustermap(df2,figsize=(4,2))
```

### Pivot Table Heat Map

```python
pvtflight=flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
pvtflight
```

```python
sns.heatmap(pvtflight)
```
