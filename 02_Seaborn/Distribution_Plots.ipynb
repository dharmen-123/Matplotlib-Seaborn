## *********** Distribution Plots ***********

######  We use distribution plots to visualize the distribution of quantitative data.
######  Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:
######  distplot/histplot
######  jointplot
######  pairplot
######  rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
print("Import successfully")
```

```python
df=sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

### Distplot

```python
sns.displot(df['total_bill'],bins=20)
```

```python
# Histplot and displot are same
sns.histplot(df['total_bill'],bins=8)
```

```python
sns.histplot(df['total_bill'],bins=20,kde=True,color='red')
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=25,kde=True,color='green')
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=25,kde=True,color='purple')
```

### JointPlot

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='scatter')
# kinds parameter types
# 'scatter' , 'kde' , 'hex', 'hist', 'reg', 'resid'
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

### Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='size',palette='dark')
```

### Rugplot

```python
sns.rugplot(df,x='total_bill',y='tip',)
```

```python
sns.rugplot(df,palette='rainbow',)
```
