## *************** MATPLOTLIB ***************

```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
print("Import Successfully")
```

```python
x=np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
plt.title("Basic Graph")
plt.xlabel("X- axis")
plt.ylabel("Y- axis")
plt.show()
```

### Subplots in matplotlib

```python
plt.subplot(2,2,1)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,y**6)
```

### Object Oriented Way

```python
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2=fig.add_axes([0.3,0.45,0.3,0.4])
axis2.plot(x,y)
```

```python
canva=plt.figure(figsize=(16,10),dpi=100)  # define the canva size and dpi= dot per inch
ax1=canva.add_axes([0.1,0.1,0.6,0.6])
ax1.plot(x-1,y**2)
ax2=canva.add_axes([0.2,0.2,0.2,0.4])
ax2.plot(x**5,y/x)
```
