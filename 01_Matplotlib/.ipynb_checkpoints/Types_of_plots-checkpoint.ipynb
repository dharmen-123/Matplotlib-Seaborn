```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
print("Import Successfully")
```

## ************** Types of Plots **************

### Scatter Plot

```python
x=np.random.randint(1,5,15)
x
```

```python
y=np.linspace(2,10,15)
y
```

```python
plt.scatter(x,y)
plt.show()
```

```python
a=np.linspace(1,5,10)
b=np.linspace(5,10,10)
b
```

```python
plt.scatter(a,b)
plt.show()
```

### Histogram

```python
plt.hist(y)
```

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

### Boxplot

```python
plt.boxplot(a)
plt.show()
```

### Saving the plots

```python
fig=plt.figure()
ax=fig.add_axes([.1,.1,.8,.8])
ax.plot(a,b**5)
plt.savefig("../Images/basicfig.png")
```
