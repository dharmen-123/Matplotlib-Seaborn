```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================================
# DATA
# ==========================================================

x = np.linspace(0, 10, 200)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2*x)
y4 = np.cos(2*x)

# ==========================================================
# FIGURE
# ==========================================================

fig, ax = plt.subplots(figsize=(15,8), dpi=150)

fig.set_facecolor("#F5F5F5")

fig.suptitle(
    "Complete Matplotlib Styling Showcase",
    fontsize=22,
    fontweight='bold',
    color='darkblue'
)

ax.set_facecolor("#FCFCFC")

# ==========================================================
# LINE 1
# ==========================================================

ax.plot(
    x,
    y1,
    label="Sin(x)",
    color="royalblue",
    linewidth=3,
    linestyle="-",
    marker="o",
    markersize=7,
    markevery=15,
    markerfacecolor="white",
    markeredgecolor="blue",
    markeredgewidth=2,
    alpha=0.9
)

# ==========================================================
# LINE 2
# ==========================================================

ax.plot(
    x,
    y2,
    label="Cos(x)",
    color="crimson",
    linewidth=2.5,
    linestyle="--",
    marker="s",
    markersize=6,
    markevery=18,
    alpha=0.8
)

# ==========================================================
# LINE 3 (Custom Dash)
# ==========================================================

line = ax.plot(
    x,
    y3,
    label="Sin(2x)",
    color="green",
    linewidth=2.5
)[0]

line.set_dashes([10,4,2,4])

# ==========================================================
# LINE 4
# ==========================================================

ax.plot(
    x,
    y4,
    label="Cos(2x)",
    color="purple",
    linewidth=2,
    linestyle=":",
    marker="D",
    markersize=5,
    markevery=20
)

# ==========================================================
# SCATTER
# ==========================================================

ax.scatter(
    x[::8],
    y1[::8],
    s=70,
    color="orange",
    edgecolor="black",
    alpha=0.8,
    label="Scatter"
)

# ==========================================================
# FILL AREA
# ==========================================================

ax.fill_between(
    x,
    y1,
    y2,
    where=(y1>y2),
    color="cyan",
    alpha=0.25,
    label="Fill Between"
)

# ==========================================================
# TITLE
# ==========================================================

ax.set_title(
    "Matplotlib Design Features",
    fontsize=18,
    fontweight="bold",
    color="darkred",
    pad=20
)

# ==========================================================
# AXIS LABELS
# ==========================================================

ax.set_xlabel(
    "X Axis",
    fontsize=14,
    fontweight="bold",
    color="navy"
)

ax.set_ylabel(
    "Y Axis",
    fontsize=14,
    fontweight="bold",
    color="navy"
)

# ==========================================================
# LIMITS
# ==========================================================

ax.set_xlim(0,10)
ax.set_ylim(-2,2)

# ==========================================================
# GRID
# ==========================================================

ax.grid(
    True,
    which="major",
    linestyle="--",
    linewidth=0.8,
    color="gray",
    alpha=0.6
)

ax.minorticks_on()

ax.grid(
    which="minor",
    linestyle=":",
    linewidth=0.4,
    alpha=0.5
)

# ==========================================================
# TICKS
# ==========================================================

ax.tick_params(
    axis='both',
    direction='inout',
    length=8,
    width=2,
    colors='darkgreen',
    labelsize=11
)

plt.xticks(rotation=20)

# ==========================================================
# SPINES
# ==========================================================

ax.spines["top"].set_color("red")
ax.spines["top"].set_linewidth(3)

ax.spines["right"].set_color("blue")
ax.spines["right"].set_linewidth(3)

ax.spines["left"].set_color("green")
ax.spines["left"].set_linewidth(3)

ax.spines["bottom"].set_color("purple")
ax.spines["bottom"].set_linewidth(3)

# ==========================================================
# LEGEND
# ==========================================================

ax.legend(
    loc="upper right",
    fontsize=11,
    title="Functions",
    title_fontsize=13,
    fancybox=True,
    shadow=True,
    framealpha=0.9,
    edgecolor="black"
)

# ==========================================================
# TEXT
# ==========================================================

ax.text(
    6.3,
    -1.6,
    "Matplotlib Styling",
    fontsize=14,
    color="darkblue",
    fontweight="bold",
    bbox=dict(
        boxstyle="round",
        facecolor="yellow",
        edgecolor="red",
        alpha=0.5
    )
)

# ==========================================================
# ANNOTATION
# ==========================================================

peak = np.argmax(y1)

ax.annotate(
    "Maximum",
    xy=(x[peak], y1[peak]),
    xytext=(3,1.6),
    fontsize=12,
    color="darkred",
    arrowprops=dict(
        arrowstyle="->",
        color="black",
        lw=2
    )
)

# ==========================================================
# HORIZONTAL & VERTICAL LINES
# ==========================================================

ax.axhline(
    y=0,
    color="black",
    linewidth=1.5
)

ax.axvline(
    x=5,
    color="gray",
    linestyle="--",
    linewidth=1.5
)

# ==========================================================
# HIGHLIGHT REGIONS
# ==========================================================

ax.axvspan(
    2,
    4,
    color="orange",
    alpha=0.2
)

ax.axhspan(
    -0.5,
    0.5,
    color="lime",
    alpha=0.15
)

# ==========================================================
# SAVE FIGURE
# ==========================================================

# plt.savefig("../Images/matplotlib_styling.png",
#             dpi=300,
#             bbox_inches="tight")

# ==========================================================
# LAYOUT
# ==========================================================

plt.tight_layout()

plt.show()
```
